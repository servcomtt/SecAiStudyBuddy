## Exercise 5: Calculate Model Accuracy Metrics

Key regression evaluation metrics:
- **MAE** (Mean Absolute Error) — average absolute prediction error; easy to interpret
- **RMSE** (Root Mean Squared Error) — penalizes large errors more than MAE

> **Exam Tip:** For classification models, know: **Accuracy, Precision, Recall, F1-score, and Confusion Matrix**. For regression, know **MAE, RMSE, and R²**. The exam distinguishes between regression and classification evaluation metrics.
