## Exercise 4: Detect Outliers with IQR and Visualize

The **Interquartile Range (IQR)** method detects outliers:
- IQR = Q3 - Q1
- Lower bound = Q1 - 1.5 × IQR
- Upper bound = Q3 + 1.5 × IQR

Values outside these bounds are considered outliers.

> **Exam Tip:** The IQR method is non-parametric (no assumption of normality). The Z-score method assumes a normal distribution. The exam may ask which is more appropriate for skewed data — IQR is the safer choice.

## Exercise 5: Min-Max Normalization

**Normalization** rescales numeric data to a standard range (typically 0–1). This prevents large-magnitude features from dominating in ML models.

Formula: `x_norm = (x - x_min) / (x_max - x_min)`

> **Exam Tip:** Know the difference between **normalization** (rescale to [0,1]) and **standardization** (Z-score, mean=0 std=1). Normalization is sensitive to outliers; standardization is more robust. The exam may present both and ask when each is appropriate.


```python
df5 = df.copy()
# Clean up first: fill NaN salary and remove the outlier age=200 for this exercise
df5['salary'].fillna(df5['salary'].median(), inplace=True)
df5 = df5[df5['age'] != 200].copy()
df5['age'].fillna(df5['age'].median(), inplace=True)

# TODO: Apply min-max normalization to the 'salary' column
#       Formula: (x - min) / (max - min)
#       Store result in a new column 'salary_normalized'

# TODO: Apply min-max normalization to the 'age' column → 'age_normalized'

# TODO: Print min, max, and range of both original and normalized columns to verify
print(df5[['salary', 'salary_normalized', 'age', 'age_normalized']].describe().round(3))
```
