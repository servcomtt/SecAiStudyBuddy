## Exercise 1: Descriptive Statistics — Manual and with NumPy

Key measures of central tendency and spread:
- **Mean** — average; sensitive to outliers
- **Median** — middle value; robust to outliers
- **Mode** — most frequent value
- **Range** — max - min
- **Variance** — average squared deviation from mean
- **Standard Deviation** — square root of variance; same units as data

> **Exam Tip:** When data is skewed, the **median** is a better measure of center than the mean. The exam often asks: in a right-skewed distribution, is mean > median or mean < median? Answer: mean > median (the tail pulls the mean right).

## Exercise 2: Histogram, Normal Distribution, and Skewness

A **histogram** shows the frequency distribution of a variable. Compare it to a normal (bell) curve to assess normality.

**Skewness:**
- Symmetric (skew ≈ 0) → mean ≈ median
- Right-skewed (positive skew) → tail on the right → mean > median

## Exercise 3: Z-Scores

A **z-score** measures how many standard deviations a value is from the mean:

`z = (x - mean) / std_dev`

## Exercise 4: One-Sample T-Test

A **t-test** determines whether a sample mean is significantly different from a known population mean (or another sample). Key concepts:
- **Null hypothesis (H₀):** No difference exists
- **Alternative hypothesis (H₁):** A difference exists
- **p-value:** Probability of observing results at least as extreme as ours, assuming H₀ is true
- **Significance level (α):** Threshold for rejecting H₀ (commonly 0.05)

> **Exam Tip:** If p-value < α (usually 0.05), **reject** the null hypothesis. If p-value ≥ α, **fail to reject** (NOT the same as accepting!) H₀. The exam loves this distinction.

## Exercise 5: Correlation Coefficient

The **Pearson correlation coefficient (r)** measures the strength and direction of a linear relationship:
- r = 1.0: perfect positive correlation
- r = -1.0: perfect negative correlation
- r = 0: no linear correlation

> **Exam Tip:** Correlation does NOT imply causation! The exam may present a correlation finding and ask what conclusions can be drawn — always choose the answer that avoids claiming causality without a controlled experiment.

## Exercise 6: Simple Linear Regression with NumPy

- **m (slope):** Change in y per unit increase in x
