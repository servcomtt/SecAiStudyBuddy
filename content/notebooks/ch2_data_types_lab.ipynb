## Exercise 1: Python Built-In Data Types

Python has several native types that map to CompTIA SecAI+ categories:
- `int` / `float` → Quantitative (discrete / continuous)
- `str` → Qualitative / Text
- `bool` → Binary / Flag
- `datetime` → Temporal

> **Exam Tip:** Timestamps and dates are a separate data type category on the exam. Always check if a date column is stored as a string vs a proper datetime object.


```python
from datetime import datetime

# TODO: Create one variable of each type below
age = None          # int
salary = None       # float
name = None         # str
is_active = None    # bool
join_date = None    # datetime

# TODO: Print each variable and its type using type()
for label, val in [("age", age), ("salary", salary), ("name", name),
                   ("is_active", is_active), ("join_date", join_date)]:
    print(f"{label}: {val}  -> type: {type(val).__name__}")
```
