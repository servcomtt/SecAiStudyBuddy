## Exercise 5: Box Plot — Distribution Comparison

**Use when:** Comparing the distribution of a numeric variable across multiple categories.

A box plot shows:
- **Box:** IQR (Q1 to Q3)
- **Line in box:** Median
- **Whiskers:** 1.5 × IQR from Q1/Q3
- **Points beyond whiskers:** Outliers

> **Exam Tip:** Box plots are better than bar charts for showing **distribution shape**. A box plot instantly reveals skewness (median off-center in box), spread (box width), and outliers (points outside whiskers). Bar charts only show a single value per category.
